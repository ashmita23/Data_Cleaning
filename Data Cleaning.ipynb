```python
food=pd.read_csv('food-consumption.csv')
food.head()
```

```python
food.isnull().sum()
```

```python
import seaborn as sns
import matplotlib as plt
import numpy as np
```

```python
sns.heatmap(food.isnull(),yticklabels=False, annot=True)
```

```python
for col in food.columns:
    missing = food[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('created missing indicator for: {}'.format(col))
        food['{}_ismissing'.format(col)] = missing


# then based on the indicator, plot the histogram of missing values
ismissing_cols = [col for col in food.columns if 'ismissing' in col]
food['num_missing'] = food[ismissing_cols].sum(axis=1)

food['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```

```python
#imputing missing values
food_numeric=food.select_dtypes(include=[np.number])
cols= food_numeric.columns.values

for col in food_numeric:
    missing =food[col].isnull()
    num_missing =np.sum(missing)
    
    if num_missing >0:
        food['{}_ismissing'.format(col)]=missing
        med= food[col].median()
        food[col]=food[col].fillna(med)
```

```python
food.boxplot(column=['Yoghurt'])
```

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
df=pd.read_csv('wine.csv')
df.head()
```

```python
df['overall acidity']= df['fixed acidity']+df['volatile acidity']
```

```python
colstodrop= ['fixed acidity','volatile acidity']
df=df.drop(colstodrop,axis=1)
```

```python
df.isnull().sum()
```

```python
#representation of the null values
colours = ['#000099', '#ffff00']
sns.heatmap(df.isnull(), cmap= sns.color_palette(colours))
```

```python
wine_numeric=df.select_dtypes(include=[np.number])
cols= wine_numeric.columns.values

for col in wine_numeric:
    missing =df[col].isnull()
    num_missing =np.sum(missing)
    
    if num_missing >0:
        df['{}_ismissing'.format(col)]=missing
        med= df[col].median()
        df[col]=df[col].fillna(med)
```

```python
df.head()
```

```python
columns=['citric acid_ismissing','residual sugar_ismissing',
       'chlorides_ismissing','pH_ismissing','sulphates_ismissing','overall acidity_ismissing']
df=df.drop(columns,axis=1)
```

```python
df.head()
```

```python
df.describe()
```

```python
df['quality'].hist(bins=100)
```

```python
#outliers
```

```python
#to calculate if the rows have 90% similarity because only then can we omit it
num_rows = len(df.index)
low_information_cols = [] #

for col in df.columns:
    cnts = df[col].value_counts(dropna=False)
    top_pct = (cnts/num_rows).iloc[0]
    
    if top_pct > 0.9:
        low_information_cols.append(col)
        print('{0}: {1:.5f}%'.format(col, top_pct*100))
        print(cnts)
        print()
#this shows that there is no similarity between the columns
```
